### OLS with Numpy

```python
x0.reshape(-1, 1).shape
```

```python
# X data generation
n_data = 200
x1 = np.linspace(200, 500, n_data)
x0 = np.ones(n_data)
X = np.hstack(( x0.reshape(-1, 1 ) , x1.reshape(-1, 1 ) ))
X.shape
```

```python
# select parameters
beta = np.array([5, -2]).reshape(-1, 1 )
beta.shape
```

```python
# y ture
y_true = X @ beta
y_true.shape
```

```python
y_true
```

```python
y_true + (np.random.normal(0, 1, n_data) * 20).reshape(-1, 1)
```

```python
#   add random normal noise
sigma = 20
y_actual = y_true + (np.random.normal(0, 1, n_data) * sigma).reshape(-1, 1)
print(y_actual[0:4, :])
```

The matrix equation for the estimated linear parameters is as below:
$${\hat {\beta }}=(X^{T}X)^{-1}X^{T}y.$$

```python
# estimations
beta_estimated = np.linalg.inv(X.T @ X) @ X.T @ y_actual
```

```python
import matplotlib.pyplot as plt

plt.plot(x1, y_actual, 'o')
plt.plot(x1, y_true, 'g-', c = 'black')
```

Calculate the sum of squared residual errors
$$
RSS=y^{T}y-y^{T}X(X^{T}X)^{{-1}}X^{T}y
$$

```python
y_actual
```

```python
RSS = ( y_actual.T @ y_actual - y_actual.T @ X @ np.linalg.inv(X.T @ X) @ X.T @ y_actual )
```

Calculated the Total Sum of Squares of the spread of the actual (noisy) values around their mean
$$
TSS=(y-{\bar  y})^{T}(y-{\bar  y})=y^{T}y-2y^{T}{\bar  y}+{\bar  y}^{T}{\bar  y}
$$

```python
y_mean = ( np.ones(n_data) * np.mean(y_actual) ).reshape( -1 , 1 )
TSS = (y_actual - y_mean).T @ (y_actual - y_mean)
TSS
```

```python
# get predictions
y_pred = X @ beta_estimated
```

Calculate the Sum of Squares of the spread of the predictions around their mean.
$$
ESS=({\hat  y}-{\bar  y})^{T}({\hat  y}-{\bar  y})={\hat  y}^{T}{\hat  y}-2{\hat  y}^{T}{\bar  y}+{\bar  y}^{T}{\bar  y}
$$

```python
ESS = (y_pred - y_mean).T @ (y_pred - y_mean)

ESS
```

```python
TSS, ESS + RSS
```

Get $R^2$
$$
1 - RSS / TSS
$$

```python
1 - RSS / TSS
```

###  Standard error of regression
Calculate the standard error of the regression. We divide by `(n-2)`, because the Expectation of the sum of squares is `(n-2)*sigma^2`.

```python
sr2 = ( (1 / (n_data - 2)) * (y_pred - y_actual).T  @ (y_pred - y_actual))
sr = np.sqrt(sr2)
sr
```

### Get variance and covariance Matrix
In order to get the standard errors for our linear parameters, we use the matrix formula below:
$$
Var(β^)=σ^2(X′X)^{-1}
$$

```python
var_beta = sr2 * np.linalg.inv(X.T @ X)
var_beta
```

```python
print(
    f'Std Error for b0 {np.sqrt(var_beta[0, 0])}, \nStd Error for b1 {np.sqrt(var_beta[1, 1])}'
)
```

```python
df3_rec_stack.set_index(  [ 'CCDD' ,'CCPP' , 'CCDI' , "STATS" ]  ).unstack().head()
```

```python
df_l_w = df3_rec_stack.pivot( index = [ 'CCDD' ,'CCPP' , 'CCDI' ], 
                         columns = 'STATS' ,
                         values = 'VALUES' 
                        ).rename_axis( [None], axis = 1 ).reset_index()
df_l_w.head()
```

### 4.1.2. If condition with more than 1 expression
We will use elif. This function allows us to add more test expressions.


if test expresion1:
    Code1

elif test expresion2:
    Code2

elif test expresion3:
    Code3

elif test expresion4:
    Code4

elif test expresion5:
    Code5

else :
    Code6

</font>

Python will read the conditions of the test in order. If Test expression 1 is True, the rest of the test expressions will not be evaluated. R execute Code2 and will not test the next conditions.

In case no Test expression is True, the Code6 will be excuted.

### Examples


```python
A = 20
B = 30

if A == B:
    print( 'Test expresion 1 is True' )
    
elif A > B:
    print( 'Test expresion 2 is True' )

elif A >= B:
    print( 'Test expresion 3 is True' )
    
elif A < B:
    print( 'Test expresion 4 is True' )
    
elif A <= B:
    print( 'Test expresion 5 is True' )
    
else:
    print( 'Test expresion 6 is True' )
```

```python
# A = 20
# B = 30

if A == B:
     print ('A is the same as B. Condition 1 is true.')
    
elif A > B:
     print ('A is greater than B. Condition 2 is true.')

elif A >= B:
     print ('A is greater than or equal to B. Condition 3 is true.')
    
elif A < B:
     print ('A is less than B. Condition 4 is true.')
    
elif A <= B:
     print ('A is less than or equal to B. Condition 5 is true.')
    
else:
     print ('When neither condition is true, this phrase is printed.')
```

### 4.3. Pass, Continue, Break, Try

Function	Description
Pass	Allows you to handle the condition
without the loop being impacted in any way
Continue	The current iteration of the loop will be disrupted,
but the program will return to the top of the loop
Break	The loop will exit
Try	This command lets you test a block of code for errors.
Exception	An exception is an event, which occurs during
the execution of a program that disrupts the normal
flow of the program's instructions.
It allows you to raise a message to help the user.

### 4.3.1 Pass

```python
for number in range(10):

    print('Number is ' + str(number))
```

```python
a = 2
if a == 1:
    print(a)
```

```python
prime = [2 , 3, 5, 7]
```

```python
for number in range(10):
    
    if (number in prime):
        # comment
        pass

    elif number == 3:
        #number = number + 1
        print( f'This number is {number}' )

    else:
        #number = number + 2
        print('Different number')

    print('Number is ' + str(number), '\n')

print('Out of loop')
```

### 4.3.2. Continue

```python
for number in range(10):
    if number == 5:
        # comment
        continue

    elif number == 3:
        #number = number + 1
        print( f'This number is {number}' )

    else:
        #number = number + 2
        print('Different number')
    
    print('Number is ' + str(number), '\n')

print('Out of loop')
```

### 4.3.3. Break

```python
for number in range(10):
    if number == 5:
        # comment
        print('This is the output of break.')
        break

    elif number == 3:
        #number = number + 1
        print( f'This number is {number}' )

    else:
        #number = number + 2
        print('Different number')

    print('Number is ' + str(number))

print('Out of loop')
```

```python
for number in range(10):
    if number == 5:
        # comment
        raise Exception('This is the output of exception.')

    elif number == 3:
        #number = number + 1
        print( f'This number is {number}' )

    else:
        #number = number + 2
        print('Different number')

    print('Number is ' + str(number))

print('Out of loop')
```
