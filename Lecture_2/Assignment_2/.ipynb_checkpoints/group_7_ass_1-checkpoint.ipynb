## Restricted Least Squares

Given the following equation:


$$
\begin{aligned} 
lnCT &= \beta_{0}+\beta_{q}lnq+ \frac{1}{2}\beta_qq lnq^2+\beta_{q1}lnqlnp_1+\beta_{q2}lnqlnp_2+ \beta_{q1}lnqlnp_{3} +\beta_{1}lnp_1+\beta_{2}lnp_2+ \beta_{3}lnp_3 \\
& + \frac{1}{2}\beta_{11}ln^{2}p_1+ \frac{1}{2}\beta_{22}ln^{2}p_{2}+ \frac{1}{2}\beta_{33}ln^{2}p_{3} + \frac{1}{2}\beta_{12}lnp_{1}lnp_{2}+ \frac{1}{2}\beta_{13}lnp_{1}lnp_{3}+\frac{1}{2}\beta_{23}lnp_{2}lnp_{3} 
\end{aligned}
$$


ST: 

<br>
$$
\begin{aligned} 
\beta_{1} + \beta_{2} + \beta_{3} &= 1 \\
\beta_{q1} + \beta_{q2} + \beta_{q3} &= 0 \\
\beta_{11} + \beta_{12} + \beta_{13} &= 0 \\
\beta_{21} + \beta_{22} + \beta_{23} &= 0 \\
\beta_{31} + \beta_{32} + \beta_{33} &= 0 \\
\beta_{ij} = \beta_{ji}
\end{aligned}
$$


1. Get $\widehat{\boldsymbol{\beta}}^{(RLS)}$ vector from the equation bellow. Use the `q`, `p1`, `p2`, `p3`, `CT` numpies. <br><br>
2. Get the `covariance matrix` $\mathbb{V}{\rm ar} \left(\widehat{\boldsymbol{\beta}}^{(RLS)} \right)$. <br><br>
**Hint: For more information about Restricted Least Squares [click here](http://web.vu.lt/mif/a.buteikis/wp-content/uploads/PE_Book/4-4-Multiple-RLS.html).**<br>

```python
import pandas as pd
import numpy as np
greene = pd.read_csv(r"D:/Usuarios/Usuario/Documents/GitHub/Diplomado_PUC/_data/christensen_greene_f4.csv")
```

```python
len(greene)
```

```python
ct = greene.COST.values.reshape(-1, 1 )
q = greene.Q.values.reshape(-1, 1 )
p1 = greene.PL.values.reshape(-1, 1 )
p2 = greene.PF.values.reshape(-1, 1 )
p3 = greene.PK.values.reshape(-1, 1 )
```

```python
lnct =( np.log(greene.COST.values) ).reshape(-1, 1 )  

lnq = ( np.log(greene.Q.values) ).reshape(-1, 1 )
lnp1 =( np.log(greene.PL.values) ).reshape(-1, 1 )
lnp2 = ( np.log(greene.PF.values) ).reshape(-1, 1 )
lnp3 = ( np.log(greene.PK.values) ).reshape(-1, 1 )

lnq2 = ( np.log(greene.Q.values**2) ).reshape(-1, 1 )
ln2p1 = ( np.log(greene.PL.values)**2 ).reshape(-1, 1 )
ln2p2 = ( np.log(greene.PF.values)**2 ).reshape(-1, 1 )
ln2p3 = ( np.log(greene.PK.values)**2 ).reshape(-1, 1 )
```

y vector

```python
y =  lnct.copy()
```

creating a vector of zeros

```python
x0 = np.ones( len(greene) ).reshape(-1, 1 )
```

creating the X matrix

```python
X = np.hstack(( x0, lnq, 0.5*(q*lnq2), (lnq*lnp1), (lnq*lnp2), (lnq*lnp3), lnp1, lnp2, lnp3,
              0.5*(ln2p1), 0.5*(ln2p2), 0.5*(ln2p3), 0.5*(lnp1*lnp2), 0.5*(lnp1*lnp3), 0.5*(lnp2*lnp3)))
```

```python
X.shape
```

Creating the matrix for the restrictions

```python
r = np.array([1,0,0,0,0]).reshape(-1, 1 )
r.shape
```

```python
L = np.zeros((5,15))
L
```

```python
L[0,6:9] = 1

L[1,3:6] = 1

L[2,9] = 1
L[2,12:14] = 1 

L[3,10] = 1
L[3,12] = 1
L[3,14] = 1 

L[4,11] = 1
L[4,13:15] = 1
```

```python
L
```

```python
L.shape
```

```python
#1
#First, we calculate the beta ols
beta_ols = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
#Then, we calculte the beta rls
beta_rls = beta_ols - np.linalg.inv(X.T @ X)  @ L.T @ np.linalg.inv(L @ np.linalg.inv(X.T @ X) @ L.T) @ (L@ beta_ols - r)
```

```python
beta_rls
```

```python
beta_rls.shape
```

```python
#2
# get predictions
y_pred = X @ beta_ols
```

```python
# calculate the standard error of the regression
sr2 = ( (1 / (len(greene) - 2)) * (y_pred - y).T  @ (y_pred - y))
sr = np.sqrt(sr2)
sr
```

```python
#Get variance and covariance Matrix
var_beta_ols = sr2 * np.linalg.inv(X.T @ X)
```

```python
#Get the covarianca matrix of beta RLS
var_beta_rls = var_beta_ols - var_beta_ols @ L.T @ np.linalg.inv( L@ np.linalg.inv(X.T @ X) @ L.T ) @ L@ np.linalg.inv(X.T @ X)
```
