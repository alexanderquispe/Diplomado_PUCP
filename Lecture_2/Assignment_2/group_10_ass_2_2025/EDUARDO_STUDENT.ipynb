```python
## PREGUNTA 6
def es_positivo(x):
    return f"The number {x} is positive." if x > 0 else f"The number {x} is not positive."

print(es_positivo(7))
print(es_positivo(-3))
```

```python
## PREGUNTA 7
def mensaje_temp_c(t):
    if t < 0:
        return "It is freezing."
    elif 0 <= t <= 20:
        return "It is cold."
    elif 21 <= t <= 30:
        return "It is warm."
    else:
        return "It is hot."

for t in [-5, 10, 25, 33]:
    print(t, "=>", mensaje_temp_c(t))
```

```python
## PREGUNTA 8
def elegible_beca(gpa, actividades_extras, horas_servicio):
    if (gpa > 3.5) and (actividades_extras == "Yes" or horas_servicio > 50):
        return "Eligible for scholarship."
    return "Not eligible for scholarship."

print(elegible_beca(3.6, "No", 60))
print(elegible_beca(3.2, "Yes", 70))
print(elegible_beca(3.8, "No", 10))
```

```python
## PREGUNTA 9
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print("list1 is list2:", list1 is list2)   # False: objetos distintos en memoria
print("list1 is list3:", list1 is list3)   # True: misma referencia
print("list1 == list2:", list1 == list2)   # True: mismo contenido (igualdad)

# Explicación:
# is  -> identidad (¿es el MISMO objeto?)
# ==  -> igualdad (¿tiene el mismo contenido?)
```

```python
## PREGUNTA 10
def calificacion(score):
    if score >= 90:
        return "A"
    elif 80 <= score <= 89:
        return "B+" if score == 85 else "B"
    elif 70 <= score <= 79:
        return "C"
    else:
        return "Fail"

for s in [95, 85, 82, 71, 60]:
    print(s, "=>", calificacion(s))
```
