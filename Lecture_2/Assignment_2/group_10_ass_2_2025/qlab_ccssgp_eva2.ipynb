6.- Basic If Condition (Hecho por Georgina)

Write a Python function that checks if a given number is positive. If the number is greater than zero, return "The number X is positive.". Otherwise, return "The number X is not positive."

```python
def calculator_square( x ):
    
    x = x
    
    result = x   
    
    if ( 0 < result ):
        print(f'The number {result} is positive')
    
    else:
        print( f'The number {result} is not positive')
    
    return result
```

7. If Condition with Multiple Expressions - Georgina
* Create a program that checks the temperature (in Celsius) and returns a message depending on the value:
  * If the temperature is below 0, return "It is freezing.".
  * If the temperature is between 0 and 20, return "It is cold.".
  * If the temperature is between 21 and 30, return "It is warm.".
  * If the temperature is greater than 30, return "It is hot.".

```python
T_Cesius = 50

if T_Cesius < 0: 
    print('It is freezing')

elif (T_Cesius > 0) and (T_Cesius <20):
    print('It is cold')

elif (T_Cesius >= 21) and (T_Cesius <30):
    print('It is warm')

else:
    print('It is hot')
```

8. Logical Operators - Georgina
* Write a function that determines if a person is eligible for a scholarship based on these conditions:
  * The person must have a GPA greater than 3.5 AND
  * Either their extracurricular activities are "Yes" OR they have community service hours     greater than 50.
 * The function should return:
    * "Eligible for scholarship." if conditions are met.
    * "Not eligible for scholarship." otherwise.

```python
def scholarship_eligibility(GPA,extra_activ, comun_serv_hour):
    if GPA > 3.5 and (extra_activ == "Yes" or comun_serv_hour > 50 ):
        return("Eligible for scholarship")
    else:
        return("Not eligible for scholarship")
```

```python
scholarship_eligibility(0,"Yes", 55)
```

```python
for num in range(1, 21):
    if num % 2 == 0:
        print(num)
```

```python
# Modificamos el loop para sumar los elementos:
total = 0   
for num in range(1, 21):
    if num % 2 == 0:
        print(num)
        total += num   
print("La suma de los números pares es:", total)
```
