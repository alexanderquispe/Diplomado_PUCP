```python
#7. If conditions with multiple expressions:

while True:
    try:
        t = float(input("Enter temperature in Celsius: "))
        break
    except ValueError:
        print("Please enter a valid number.")

if t < 0:
    print("It is freezing.")
elif 0 <= t <= 20:
    print("It is cold.")
elif 21 <= t <= 30:
    print("It is warm.")
else:
    print("It is hot.")
```

```python
#8. Logical operators

while True:
    try:
        gpa = float(input("GPA (e.g., 3.6): "))
        break
    except ValueError:
        print("Please enter a valid number for GPA.")

# Extracurricular
extracurricular = input('Extracurricular activities? (Yes/No): ').strip()

# Community service hours
while True:
    try:
        hours = int(input("Community service hours (integer): "))
        break
    except ValueError:
        print("Please enter a whole number for hours.")

if gpa > 3.5 and (extracurricular == "Yes" or hours > 50):
    print("Eligible for scholarship.")
else:
    print("Not eligible for scholarship.")
    
```

```python
#9. Python Identity Operators ---
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print("list1 is list2:", list1 is list2)   # False: objetos distintos en memoria
print("list1 is list3:", list1 is list3)   # True: misma referencia/objeto
print("list1 == list2:", list1 == list2)   # True: contenidos iguales

print("\nExplanation:")
print("- list1 is list2 → False: same values, but different memory locations (different objects).")
print("- list1 is list3 → True: list3 points to exactly the same object as list1.")
print("- list1 == list2 → True: their contents are equal, even if they are different objects.")
```

```python
#10. Nested If Statement:
while True:
    try:
        score = int(input("Enter student score (0-100): "))
        if 0 <= score <= 100:
            break
        else:
            print("Score must be between 0 and 100.")
    except ValueError:
        print("Please enter a whole number between 0 and 100.")

if score >= 90:
    print("A")
elif 80 <= score <= 89:
    if score == 85:
        print("B+")
    else:
        print("B")
elif 70 <= score <= 79:
    print("C")
else:
    print("Fail")
```
