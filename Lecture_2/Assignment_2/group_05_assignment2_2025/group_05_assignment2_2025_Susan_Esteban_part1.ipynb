### Change the data type of a variable

```python
import numpy as np
```

```python
df_2_sub = df_2[['CONGLOME','VIVIENDA','HOGAR','CODPERSO','MES','SUB_CONGLOME','UBIGEO']].copy()
```

```python
df_2_sub.loc[:, 'UBIGEO'] = df_2_sub['UBIGEO'].astype(str).str.zfill(6)
df_2_sub['UBIGEO'].head()
```

```python
df_2_sub.loc[:, 'UBIGEO'] = df_2_sub['UBIGEO'].replace({'000000': np.nan, '999999': np.nan})
df_2_sub['UBIGEO'].head()
```

```python
df_2_sub.loc[:, 'MES'] = df_2_sub['MES'].replace({0: np.nan, 13: np.nan})
df_2_sub['MES'].head()
```

###  Modify some values in a specific column.

```python
common_columns = ['CONGLOME','VIVIENDA','HOGAR','CODPERSO']
```

```python
subset = Enaho01A[['CONGLOME','VIVIENDA','HOGAR','CODPERSO','MES','SUB_CONGLOME','UBIGEO']].copy()
```

```python
subset[common_columns]  = subset[common_columns].astype('int64')
```

```python
subset.loc[:, 'UBIGEO'] = subset['UBIGEO'].astype(str).str.zfill(6)
df_2_sub.loc[:, 'UBIGEO'] = df_2_sub['UBIGEO'].astype(str).str.zfill(6)
```

## Merging Datasets

### V erify whether the values match

```python
keys_300 = set(map(tuple, subset[common_columns].values))
keys_200 = set(map(tuple, df_2_sub[common_columns].values))

print("Claves en 300:", len(keys_300))
print("Claves en 200:", len(keys_200))
print("Intersección:", len(keys_300 & keys_200))
```

### Merge

```python
df_merged = subset.merge(df_2_sub, on=common_columns, how='inner', suffixes=('_300','_200'))
df_merged.head()
```

## In the resulting DataFrame

### Group the data

```python
[c for c in df_merged.columns if 'MES' in c]
```

```python
iguales = (df_merged['MES_300'] == df_merged['MES_200']).all()
print("MES_300 y MES_200 iguales?:", iguales)
```

```python
df_merged['MES'] = df_merged['MES_300']
```

```python
if 'MES' not in df_merged.columns:
    if 'MES_300' in df_merged.columns:
        df_merged = df_merged.rename(columns={'MES_300':'MES'})
    elif 'MES_200' in df_merged.columns:
        df_merged = df_merged.rename(columns={'MES_200':'MES'})
```
