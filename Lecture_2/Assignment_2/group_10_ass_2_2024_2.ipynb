```python
#6.

# First, we use map to apply a lambda function to each email in the list emails.
# The lambda function slices each email up to (but not including) the first occurrence of '.'.
# If '.' is found, email.find('.') returns the index where '.' is located, and the slicing takes the substring before it.
# If '.' is not found, email.find('.') returns -1, and the entire email string is returned.
before_dot = list(map(lambda email: email[:email.find('.')], emails))

# The second step is to use map to apply a lambda function to each string in before_dot.
# The lambda function uses s.find('@') != -1 to check if '@' is present in the substring before the first '.'.
# If '@' is found, s.find('@') returns a non-negative index, making s.find('@') != -1 True.
# If '@' is not found, s.find('@') returns -1, making s.find('@') != -1 False.
# The map function returns a list of boolean values.
# sum(...) counts the number of True values in the list, representing the count of substrings containing '@'.
contains_at_count = sum(map(lambda s: s.find('@') != -1, before_dot))

print(contains_at_count)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1.

# When importing the data file, we have to go back twice from our current location, so our relative path needs two ../
sysdif_df = pd.read_csv("../../_data/new_up_data_sysdif_block_lvl_ver2.csv")
sysdif_df
```

```python
#2.

# Again, we need to use ../ twice
climate_df = pd.read_csv("../../_data/ESA_CLIMATE_CHANGE_tree_layers.csv")
climate_df
```

```python
#3.

# We merge both data frames using the key variables, which we specify inside on=[]
merged_df = pd.merge(sysdif_df, climate_df, on=['pc11_s_id', 'pc11_d_id', 'pc11_sd_id', 'year'])
merged_df
```

```python
#4.

# We select the specific columns to calculate descriptive statistics
# The describe() method generates summary statistics and with round(2) we round the results to 2 decimal places
stats = merged_df[['per_TREES-BD', 'per_TREES-BE', 'per_TREES-ND', 'per_TREES-NE']].describe().round(2)

print(stats)
```

```python
#5.

# Creating a new column 'sum_tree' which is the sum of the specified columns
merged_df['sum_tree'] = merged_df[["per_TREES-BD", "per_TREES-BE", "per_TREES-ND", "per_TREES-NE"]].sum(axis=1)

# Plot histogram for 'sum_tree'
plt.hist(merged_df['sum_tree'], bins=30, edgecolor='k')
plt.title('Histogram of Sum of Tree Columns')
plt.xlabel('Sum of Trees')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6.

# First, we calculate the median value of the 'sum_tree' column from the merged_df DataFrame.
median_value = merged_df['sum_tree'].median()

# Let's filter the DataFrame into two new DataFrames based on the median value.
# We create above_median_df to include only the rows where 'sum_tree' is greater than the median.
above_median_df = merged_df[merged_df['sum_tree'] > median_value]

# Similarly, we create below_median_df for rows where 'sum_tree' is less than or equal to the median.
below_median_df = merged_df[merged_df['sum_tree'] <= median_value]

# Now, we'll specify the columns we're interested in plotting.
columns_to_plot = ['per_TREES-BD', 'per_TREES-BE', 'per_TREES-ND', 'per_TREES-NE']

# We'll also define the colors for each plot.
# Colors for below the median will be lighter shades.
colors_below = ['lightskyblue', 'limegreen', 'lightsteelblue', 'mediumaquamarine']

# Colors for above the median will be darker shades.
colors_above = ['royalblue', 'green', 'slategray', 'teal']

# Let's set the figure's background color to whitesmoke and the grid lines to gainsboro for a soft appearance.
plt.rcParams['axes.facecolor'] = 'whitesmoke'
plt.rcParams['grid.color'] = 'gainsboro'

# We create a figure with 2 rows and 4 columns of subplots, giving us space for all the histograms.
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Now, we'll generate histograms for the 'Above Median' group.
# For each subplot, we plot the corresponding column, setting the color and edge color.
# We also add titles and enable grid lines for clarity.
axes[0, 0].hist(above_median_df['per_TREES-BD'], color=colors_above[0], edgecolor='black')
axes[0, 0].set_title('per_TREES-BD (Above Median)', fontweight='bold')
axes[0, 0].grid(True)

axes[0, 1].hist(above_median_df['per_TREES-BE'], color=colors_above[1], edgecolor='black')
axes[0, 1].set_title('per_TREES-BE (Above Median)', fontweight='bold')
axes[0, 1].grid(True)

axes[0, 2].hist(above_median_df['per_TREES-ND'], color=colors_above[2], edgecolor='black')
axes[0, 2].set_title('per_TREES-ND (Above Median)', fontweight='bold')
axes[0, 2].grid(True)

axes[0, 3].hist(above_median_df['per_TREES-NE'], color=colors_above[3], edgecolor='black')
axes[0, 3].set_title('per_TREES-NE (Above Median)', fontweight='bold')
axes[0, 3].grid(True)

# Next, we generate histograms for the 'Below Median' group, following the same process as above.
axes[1, 0].hist(below_median_df['per_TREES-BD'], color=colors_below[0], edgecolor='black')
axes[1, 0].set_title('per_TREES-BD (Below Median)', fontweight='bold')
axes[1, 0].grid(True)

axes[1, 1].hist(below_median_df['per_TREES-BE'], color=colors_below[1], edgecolor='black')
axes[1, 1].set_title('per_TREES-BE (Below Median)', fontweight='bold')
axes[1, 1].grid(True)

axes[1, 2].hist(below_median_df['per_TREES-ND'], color=colors_below[2], edgecolor='black')
axes[1, 2].set_title('per_TREES-ND (Below Median)', fontweight='bold')
axes[1, 2].grid(True)

axes[1, 3].hist(below_median_df['per_TREES-NE'], color=colors_below[3], edgecolor='black')
axes[1, 3].set_title('per_TREES-NE (Below Median)', fontweight='bold')
axes[1, 3].grid(True)

# Finally, we use tight_layout to adjust the spacing of the plots and show the figure.
plt.tight_layout()
plt.show()
```
