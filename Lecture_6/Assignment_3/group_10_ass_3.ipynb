```python
# Beta_OLS function
def beta_OLS( X : pd.DataFrame, y : pd.Series, intercept = True ):
        
    if not isinstance( X, pd.DataFrame ):
        raise TypeError( 'X is not a pd.DataFrame' )
    
    if not isinstance( y, pd.Series ):
        raise TypeError( 'y is not a pd.Series' )
    
    if not isinstance( intercept, bool ):
        raise TypeError( 'intercept is not a bool' )
        
    if len( X ) != len( y ):
        raise ValueError( 'X and y do not have the same number of rows' )
    
    X_mat = X.copy()
    y_vec = y.copy()
    
    if intercept == True:
        const = np.ones( len( X_mat ) )
        X_mat.insert(0, "Intercept", const)
    
    var_names = list( X_mat.columns )
    X_mat = X_mat.to_numpy()
    y_vec = y_vec.to_numpy()
    
    xtx_inv = np.linalg.inv( np.dot( np.transpose( X_mat ), X_mat ) )
    coefs = np.dot( xtx_inv, np.dot( np.transpose( X_mat ), y_vec ) )
    coefs = pd.DataFrame( coefs, index = var_names, columns = [ 'Coef.' ] )
    
    return coefs
```

```python
# var_OLS function
def var_OLS( X : pd.DataFrame, y : pd.Series, intercept = True ):
        
    if not isinstance( X, pd.DataFrame ):
        raise TypeError( 'X is not a pd.DataFrame' )
    
    if not isinstance( y, pd.Series ):
        raise TypeError( 'y is not a pd.Series' )
    
    if not isinstance( intercept, bool ):
        raise TypeError( 'intercept is not a bool' )
        
    if len( X ) != len( y ):
        raise ValueError( 'X and y do not have the same number of rows' )
    
    X_mat = X.copy()
    y_vec = y.copy()
    
    if intercept == True:
        const = np.ones( len( X_mat ) )
        X_mat.insert(0, "Intercept", const)
    
    var_names = list( X_mat.columns )
    X_mat = X_mat.to_numpy()
    y_vec = y_vec.to_numpy()
    
    xtx_inv = np.linalg.inv( np.dot( np.transpose( X_mat ), X_mat ) )
    coefs = np.dot( xtx_inv, np.dot( np.transpose( X_mat ), y_vec ) )
    
    errores=np.add(y_vec,-np.dot( X_mat , coefs ))
    var_error=np.dot(np.transpose(errores),errores)/(len(errores)-14-1)

    var_coef=var_error*xtx_inv
    
    var_coef = pd.DataFrame( var_coef, index = var_names, columns=var_names)

    
    return var_coef
```

```python
def restricted_OLS( x : pd.DataFrame, y : pd.Series, L : np.array, r: np.array,intercept = True ):
    
    if not isinstance( x, pd.DataFrame ):
        raise TypeError( 'x is not a pd.DataFrame' )
    
    if not isinstance( y, pd.Series ):
        raise TypeError( 'y is not a pd.Series' )
    
    if not isinstance( intercept, bool ):
        raise TypeError( 'intercept is not a bool' )

    if not isinstance( L, np.ndarray ):
        raise TypeError( "L is not a np.array.")  

    if not isinstance( r, np.ndarray ):
        raise TypeError( "r is not a np.array.")
     
    X_mat = x.copy()
    y_vec = y.copy()
    
    
    if intercept == True:
        #Names of variables
        var_names = list( X_mat.columns )
        var_names.insert(0,"Intercept")
        #
        const = np.ones( len( y_vec ) )
        X_mat_rs = np.c_[const,X_mat]
        #
        L_0      = np.zeros(len(r))
        L_rs        = np.c_[L_0, L]
        #
        beta = beta_OLS( x, y, intercept = True ).to_numpy()
        var = var_OLS( x, y, intercept = True ).to_numpy()       
    
        
    if intercept == False:        
        beta = beta_OLS( x, y, intercept = False ).to_numpy()
        var = var_OLS( x, y, intercept = False ).to_numpy()  
        L_rs=L.to_numpy()  
        #Names of variables
        var_names = list( X_mat.columns )
        
        
    #Declarate var as vector and matrix
    #X_mat = X_mat.to_numpy()
    y_vec = y_vec.to_numpy()
    ########################    
    
    #
    xtx_inv = np.linalg.inv( np.dot( np.transpose( X_mat_rs ), X_mat_rs ) )
    coefs = np.dot( xtx_inv, np.dot( np.transpose( X_mat_rs ), y_vec ) )
    
    #restricted_OLS
    coef_restricted_OLS= beta - xtx_inv @ L_rs.T @ np.linalg.inv(L_rs @ xtx_inv @ L_rs.T) @ (L_rs @ beta -r)
    coef_restricted_OLS = pd.DataFrame( coef_restricted_OLS, index = var_names, columns = [ 'Coef.' ] )
    coef_restricted_OLS_2=coef_restricted_OLS
    #var_restricted_OLS
    
    # Defining errors and variance
    errores=np.add(y_vec,-np.dot( X_mat_rs , coefs ))
    var_sqr = np.dot(np.transpose(errores),errores) / (len(y_vec)-len(X_mat_rs[0])-1)

    # getting var_beta_rls
    var_restricted_OLS= var_sqr * xtx_inv - var_sqr * xtx_inv @ L_rs.T @ np.linalg.inv(L_rs @ xtx_inv @ L_rs.T) @  L_rs @ xtx_inv
    var_restricted_OLS= pd.DataFrame( var_restricted_OLS, index = var_names, columns=var_names)
    
    #
    reg_OLS = pd.DataFrame( coef_restricted_OLS_2 )
        
    var_cov = var_restricted_OLS.to_numpy() 
        
    diagonal = [var_cov[i][i] for i in range( len(var_cov[0]) )] #extract diagonal
        
    std_error = np.sqrt( np.array(diagonal) )

    reg_OLS[ 'Std.Err.' ] = std_error 

    reg_OLS[ 't' ] = reg_OLS[ 'Coef.' ] / reg_OLS[ 'Std.Err.' ]
        
    p = stats.t.sf( abs( reg_OLS[ 't' ] ), df = len( y ) - len( coef_restricted_OLS ) ) * 2
        
    reg_OLS[ 'P>|t|' ] = p
                
    reg_OLS['[0.025'] = reg_OLS[ 'Coef.' ] - stats.t.ppf(q = 1 - 0.05 / 2, df = len(y)-len(coef_restricted_OLS)) * reg_OLS[ 'Std.Err.' ]
               
    reg_OLS['0.975]'] = reg_OLS[ 'Coef.' ] + stats.t.ppf(q = 1 - 0.05 / 2, df = len(y)-len(coef_restricted_OLS)) * reg_OLS[ 'Std.Err.' ]

    reg_OLS_r=reg_OLS
    
    return coef_restricted_OLS, var_restricted_OLS, reg_OLS_r
```

```python
class RegClass( object ):
    
    
    def __init__( self, x : pd.DataFrame, y : pd.Series, L: np.ndarray , r: np.ndarray, intercept = True ):
        
        __slots__ = [ 'x', 'y', 'intercept', 'beta_OLS', 'var_OLS', 'coef_restricted_OLS', 'var_restricted_OLS', ]
        
        if not isinstance( x, pd.DataFrame ):
            raise TypeError( 'x is not a pd.DataFrame' )
    
        if not isinstance( y, pd.Series ):
            raise TypeError( 'y is not a pd.Series' )
    
        if not isinstance( intercept, bool ):
            raise TypeError( 'intercept is not a bool' )
        
        if len( x ) != len( y ):
            raise ValueError( 'x and y do not have the same number of rows' )
        
        self.x = x
        self.y = y
        self.L = L
        self.r = r
        self.intercept = intercept        
   

    def reg_OLS( self ):
        
        self.beta_OLS = beta_OLS( self.x, self.y, intercept = self.intercept)
        self.var_OLS = var_OLS( self.x, self.y, intercept = self.intercept)
    
        return self.beta_OLS, self.var_OLS    
        
    def restricted_OLS( self, L : np.ndarray, r : np.ndarray, ):
      
        self.coef_restricted_OLS, self.var_restricted_OLS, reg_OLS_r=restricted_OLS(self.x, self.y, self.L, self.r, intercept = self.intercept )
        
        return self.coef_restricted_OLS, self.var_restricted_OLS, reg_OLS_r
    
    
```
