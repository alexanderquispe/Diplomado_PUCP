7. Use endes_2019 data to generate a new object named mean_key_vars to find the mean of total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715), and Age at first marriage (V511) by year and department (V024). Name these columns as mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage, respectively. Hint: Use groupby and this link.

```python
#Select the columns to group our variables by 
group_columns = ['year', 'V024']

#Select the variables to work with 
key_vars = ['V201', 'V613', 'V715', 'V511']

#group by group_colums and calculate the mean of our variables
mean_key_vars = endes_2019.groupby(group_columns)[key_vars].mean()

# Rename the columns
mean_key_vars.columns = ['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']

# Reset the index to obtain a flat DataFrame
mean_key_vars.reset_index(inplace=True)

#Call our new object
mean_key_vars


#Recoding variable V024
departamento_mapping = {
    1: 'Amazonas',
    2: 'Ancash',
    3: 'Apurimac',
    4: 'Arequipa',
    5: 'Ayacucho',
    6: 'Cajamarca',
    7: 'Callao',
    8: 'Cusco',
    9:'Huancavelica',
    10:'Huanuco',
    11:'Ica',
    12:'Junin',
    13:'La Libertad',
    14:'Lambayeque',
    15:'Lima',
    16:'Loreto',
    17:'Madre de Dios',
    18:'Moquegua',
    19:'Pasco',
    20:'Piura',
    21:'Puno',
    22:'San Martin',
    23:'Tacna',
    24:'Tumbes',
    25:'Ucayali'
}

# Apply the recodification 
mean_key_vars['V024'] = mean_key_vars['V024'].map(departamento_mapping)

#view mean_key_vars
mean_key_vars
```

8. Reshape mean_key_vars from wide to long. We want a dataframe with three columns dpto, variables, values. Name this object as reshape_mean_key_vars. Hint: Use melt method.

```python
# Reshape mean_key_vars from wide to long
reshape_mean_key_vars = mean_key_vars.melt(id_vars=['year', 'V024'],
                                           value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'],
                                           var_name='variables',
                                           value_name='values')

# Variable V024 changes to 'dpto'
reshape_mean_key_vars = reshape_mean_key_vars.rename(columns={'V024': 'dpto'})

# View the new dataframe
reshape_mean_key_vars
```
