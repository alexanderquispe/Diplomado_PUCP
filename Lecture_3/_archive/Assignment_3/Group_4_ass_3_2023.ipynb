#### 7. Use endes_2019 data to generate a new object named mean_key_vars to find the mean of total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715), and Age at first marriage (V511) by year and department (V024). Name these columns as mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage, respectively. 

```python
#Se seleccionan las variables de interés
endes_2019_new = endes_2019[['V024', 'V201', 'V613', 'V715', 'V511']]
endes_2019_new
```

```python
key_vars = ['V201', 'V613', 'V715', 'V511']

#Se vuelbven numéricas a las variables de interés. Además, se usa errors="coerce" para los NA.
endes_2019_new[key_vars] = endes_2019_new[key_vars].apply(pd.to_numeric, errors='coerce')

endes_2019_new
```

```python
# Reemplazar los códigos con etiquetas en 'V024'
endes_2019_new['department'] = endes_2019_new['V024'].map(value_labels['V024'])

#Eliminamos la columna V024
endes_2019_new = endes_2019_new.drop('V024', axis=1)

endes_2019_new = endes_2019_new[['department', 'V201', 'V613', 'V715', 'V511']]

endes_2019_new
```

```python
#Dado que la BD solo contine información del 2019, se realiza el groupby solo por departamento.
mean_key_vars = endes_2019_new.groupby(['department'])[key_vars].mean().reset_index()
mean_key_vars
```

```python
#Renombramos las columnas
mean_key_vars.columns = ['department', 'mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']
mean_key_vars
```

#### 8. Reshape mean_key_vars from wide to long. We want a dataframe with three columns dpto, variables, values. Name this object as reshape_mean_key_vars. Hint: Use melt method.

```python
#Se usa la función melt para la transformación del dataframe
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars=['department'], value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'], var_name='variables', value_name='values')
reshape_mean_key_vars
```

#### 9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible.

```python
reshape_mean_key_vars = (
    pd.melt(endes_2019[['V024', 'V201', 'V613', 'V715', 'V511']]
            .apply(pd.to_numeric, errors='coerce')
            .assign(department=endes_2019['V024'].map(value_labels['V024']))
            .drop('V024', axis=1)
            .groupby(['department'])[['V201', 'V613', 'V715', 'V511']]
            .mean()
            .reset_index()
            .rename(columns={'V201': 'mean_total_children', 'V613': 'mean_ideal_children', 'V715': 'mean_hb_yr_educ', 'V511': 'mean_first_marriage'}),
            id_vars=['department'],
            value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'],
            var_name='variable',
            value_name='value'))

reshape_mean_key_vars
```

#### 10.  Merge reshape_mean_key_vars with endes_2019. Name this object final_result. Hint: Use merge.

```python
#creamos la variable department en endes_2019 para poder hacer merge entre las bases.
endes_2019['department'] = endes_2019['V024'].map(value_labels['V024'])

#Merge
final_result = pd.merge(endes_2019, reshape_mean_key_vars, left_on=['department'], right_on=['department'])
```
