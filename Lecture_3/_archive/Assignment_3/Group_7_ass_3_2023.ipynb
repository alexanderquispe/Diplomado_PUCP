```python
# To obtain the min, max, sd, n_obs and n_missing of the indicated variables, we follow the next steps

summary_stats = (
    endes_2019
    .loc[:,['V201','V613','V715','V511']] # Keep the variables needed
    .describe() # Use 'describe' function to get the min, max, sd, n_obs and percentiles
    .loc[['count','mean','std','min','max'],:] # Keep rows that we need
    .round(2) # Round the values at 2 decimals
    .reset_index() # Use 'reset_index' function to get the index of the created data frame as a column called 'index'
)
summary_stats
```

```python
# At this point, we have a data frame with five columns (index, V201, V613, V715 and V511) and five rows where
# values contain count, mean, std, min and max. To get what is requested, we use 'pd.melt' to obtain 3 columns in 
# total (index, variables and values), then use 'pivot' to obtain a data frame with six columns (Variables, 
# count, max, mean, min and std) where the previous column 'Variables' now is the index.

p_summary_stats = (
    pd.melt(summary_stats, id_vars=['index'], var_name=['Variables'])
    .pivot(index= 'Variables',columns='index', values='value')
)
p_summary_stats
```

```python
# By the other side, to obtain the total missing values per variable, we use 'isnull()' and 'sum()' functions.

missing_counts = (
    endes_2019
    .loc[:,['V201','V613','V715','V511']] 
    .isnull() # Returns a boolean object of the same size as the input object (T = null)
    .sum() # Sum the results (T = 1, F = 0)
    .to_frame() # Convert to a frame the data
    .reset_index()
    .rename(columns = {'index':'Variables', 0:'N_missing'}) # Rename columns
)
missing_counts
```

```python
# To get the 'summary table' we use 'pd.merge' and 'sort_values' functions.

summary_table = (
    pd.merge(p_summary_stats, missing_counts, on='Variables')
    .sort_values(by='N_missing', ascending=False) # Sort by columns n_missing
    .astype({'count': int})  # Convert count as a int variable
    .rename(columns = {'count':'N_obs', 'std':'STD', 'max':'Max', 'mean':'Mean', 'min': 'Min'}) # Rename columns
    [['Variables','Min','Max','Mean','STD','N_obs','N_missing']] # Reorder
)
summary_table
```

```python
# To find the mean of total chindren ever born, ideal number of children, husbands education single years 
# and age at first marriage by year and department we use 'groupby()' function and 'agg()' as it's shown 
# in the following lines

mean_key_vars = (
    endes_2019
    .loc[:,['year','V024','V201','V613','V715','V511']] # Keep the variables needed
    .groupby( ['year','V024'], as_index = False ) # Group by 'year' and 'V024' (which contains Departments)
    .agg(Mean_total_children = ( 'V201', np.mean ), # Create new variables as the mean of other
         Mean_ideal_children = ( 'V613', np.mean ),
         Mean_hb_yr_educ = ( 'V715', np.mean ),
         Mean_first_marriage = ( 'V511', np.mean ))
    .round(2) # Round the values at 2 decimals
    .rename(columns = {'year':'Year', 'V024':'Dpto'})
    .astype({'Dpto': int}) # Convert Dpto to int type
)
mean_key_vars
```

**Exercise 8.**

Done by Mayra Gavilan Castañeda

```python
# To reshape a data frame from wide to long using melt method, we drop 'Year' variable and use 'pd.melt' function
# where the new variable 'Variables' will contain the columns of 'mean_key_vars', also it will create a column
# called 'value'.

reshape_mean_key_vars = (
    pd.melt(mean_key_vars.drop('Year', axis=1), id_vars=['Dpto'], var_name=['Variables'])
    .rename(columns = {'value':'Values'})
)
reshape_mean_key_vars
```
