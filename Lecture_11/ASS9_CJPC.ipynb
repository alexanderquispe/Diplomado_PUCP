### Part 3

#### 3.1 Generate the ratio of applicants by vacancies and store it in a new columns called ratio_applicants_vacancies (total applicants / total vacancies). Then, generate a cluster map with two groups of marker clusters. The first group should be compound by those institutes where there were more applicants than vacancies. Name this group as Applicant Surplus. The second group should be compound by those institutes where there were more vacancies than applicants. Name this group as Applicant Deficit.

```python
import folium
from folium import plugins

# Crear un mapa de Folium
a = folium.Map(location=[-12.0464, -77.0428], zoom_start=5)

# Crear la nueva columna 'ratio_applicants_vacancies'
tot_dpt_shp['ratio_applicants_vacancies'] = tot_dpt_shp['cuentadeid_postulante_procesoadm_y'] / tot_dpt_shp['ltimoden_metaatencion_y']

# Crear dos grupos de clusters: Applicant Surplus y Applicant Deficit
applicant_surplus_cluster = plugins.MarkerCluster(name='Applicant Surplus').add_to(a)
applicant_deficit_cluster = plugins.MarkerCluster(name='Applicant Deficit').add_to(a)

# Iterar sobre cada fila del DataFrame
for index, row in tot_dpt_shp.iterrows():
    # Crear un marcador en las coordenadas especificadas por LATITUD y LONGITUD de cada fila
    marker = folium.Marker([row['LATITUD'], row['LONGITUD']],
                           popup=folium.Popup(
                               # Crear una tabla HTML con informaciÃ³n de la fila, incluyendo el nuevo ratio
                               f"<table style='width:100%'><tr><th>Departamento</th><td>{row['NOMBDEP']}</td></tr>"
                               f"<tr><th>Total vancancies</th><td>{row['ltimoden_metaatencion_y']}</td></tr>"
                               f"<tr><th>Total applicants</th><td>{row['cuentadeid_postulante_procesoadm_y']}</td></tr>"
                               f"<tr><th>Total entrants</th><td>{row['sumaden_flagingresante_y']}</td></tr>"
                               f"<tr><th>Total enrolled</th><td>{row['sumaden_flagmatriculado_y']}</td></tr>"
                               f"<tr><th>Ratio Applicants/Vacancies</th><td>{row['ratio_applicants_vacancies']}</td></tr>"
                               f"</table>", max_width=300),
                           radius=50,
                           icon=folium.Icon(color="red", icon="info-sign"),
                           color="#3186cc",
                           fill=True,
                           fill_color="#3186cc",
                           tooltip="Click me!"
                           )

    # AÃ±adir el marcador al grupo correspondiente segÃºn la condiciÃ³n
    if row['cuentadeid_postulante_procesoadm_y'] > row['ltimoden_metaatencion_y']:
        marker.add_to(applicant_surplus_cluster)
    else:
        marker.add_to(applicant_deficit_cluster)

# AÃ±adir el control de capas (Layer Control) al mapa
folium.LayerControl().add_to(a)

a
```

### Part 4

#### 4.1 Generate a Circle map and plot only institutes with more applicants than vacancies. Make the radius= ratio_applicants_vacancies * 10000 

```python
import folium
from folium import plugins

# Create a map using folium
a = folium.Map(location=[-12.0464, -77.0428], zoom_start=5)

# Create a FeatureGroup for Applicant Surplus circles
applicant_surplus_circles = folium.map.FeatureGroup(name='Applicant Surplus').add_to(a)

# Create circles for institutes with more applicants than vacancies
for index, row in tot_dpt_shp.iterrows():
    if row['cuentadeid_postulante_procesoadm_x'] > row['ltimoden_metaatencion_x']:
        # Calculate the radius based on the ratio
        radius = row['ratio_applicants_vacancies'] = row['cuentadeid_postulante_procesoadm_x'] / row['ltimoden_metaatencion_x']*10000

        # Create a Circle with green color
        circle_marker = folium.Circle(location=[row['LATITUD'], row['LONGITUD']],
                                      radius=radius,
                                      color='green',
                                      fill=True,
                                      fill_color='green',
                                      fill_opacity=0.4,
                                      popup=folium.Popup(
                                          f"<strong>{row['NOMBDEP']}</strong><br>"
                                          f"Total Vacancies: {row['ltimoden_metaatencion_x']}<br>"
                                          f"Total Applicants: {row['cuentadeid_postulante_procesoadm_x']}<br>"
                                          f"Ratio Applicants/Vacancies: {row['ratio_applicants_vacancies']:.2f}"
                                      )
                                      )

        # Add the Circle to the FeatureGroup
        circle_marker.add_to(applicant_surplus_circles)

# Add Layer Control to the map
folium.LayerControl().add_to(a)

# Save the map as an HTML file or display it
a.save("applicant_surplus_map.html")

a
```
