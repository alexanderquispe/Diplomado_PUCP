```python
# Define file path
file_path = '../../_data/institutos1.csv'

# Load the data
inst = pd.read_csv(file_path)

# Show first rows of the df
inst.head()
```

```python
inst_geo = gpd.GeoDataFrame(inst, crs = "EPSG:4326",
                               geometry = gpd.points_from_xy(inst.nlong_ie,
                                                            inst.nlat_ie))

shp_dpt = gpd.read_file(r"../../_data/INEI_LIMITE_DEPARTAMENTAL/INEI_LIMITE_DEPARTAMENTAL.shp")

intersct_inst_geo = gpd.overlay(inst_geo, shp_dpt, how = "intersection")

# Variables of interest

vars_sum = ['ltimoden_metaatencion' , 'cuentadeid_postulante_procesoadm', 'sumaden_flagingresante', 'sumaden_flagmatriculado']

tot_dpt = intersct_inst_geo.groupby(['CCDD'], as_index = False )[vars_sum].sum()
tot_dpt_shp = shp_dpt.merge(tot_dpt, on = 'CCDD')
```

```python
institutos = list(zip(inst['nlat_ie'], inst['nlong_ie']))

lat_hos = inst["nlat_ie"].mean()
long_hos = inst["nlong_ie"].mean()
zoom_start = 6

instloc = fm.Map(location = [lat_hos, long_hos], tiles='cartodbpositron', zoom_start = zoom_start)

fm.Choropleth(
    geo_data=tot_dpt_shp,
    data=tot_dpt_shp,
    columns=['NOMBDEP', 'ltimoden_metaatencion'],
    key_on="feature.properties.NOMBDEP",
    fill_color="YlOrRd",
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name="Total vacancies",
    Highlight= True,
    line_color = "#000000",
    overlay=True,
    nan_fill_color = "grey"  
    ).add_to(inst)

#we use the institutos list to create a cluster marker
MarkerCluster( institutos, name = 'Cluster' ).add_to(inst)

fm.LayerControl().add_to(inst)

instloc

```
