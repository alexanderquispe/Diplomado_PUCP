```python
#  Import libraries: os library is used for file path operations, and pandas for handling dataframes
import os
import pandas as pd
```

```python
# Define the function import_data_sbs_banks with parameters for the start and end months and years.
# Aditionally, we add the end_month and end_year parameters. 
def import_data_sbs_banks(start_month, start_year, end_month=None, end_year=None):
    path = r'../../_data/sbs/B_RawData/bancos'  # Path of the directory
    found_files = []  # List for storing found files

    # Convert start_month and start_year to integers
    start_month = int(start_month)
    start_year = int(start_year)
    
    if end_month is not None:
        end_month = int(end_month)
    if end_year is not None:
        end_year = int(end_year)

    # Just in case, set end_month and end_year by default
    if end_month is None:
        end_month = start_month
    if end_year is None:
        end_year = start_year

    # Iterating over the years
    for year in range(start_year, end_year + 1):
        
        # Determine the months to be imported for the current year
        month_start = start_month if year == start_year else 1
        month_end = end_month if year == end_year else 12
        
        # Iterate over the months
        for month in range(month_start, month_end + 1):

            for day in range(1, 32):  
                # Construct file name and full path
                file_name = f'table_clean_{day}_{month}_{year}.xlsx'  # Name format
                full_path = os.path.join(path, file_name)
                
                # Check if the file exists
                if os.path.isfile(full_path):
                    found_files.append(full_path)  # Add the found file to the list

    # Print found files
    if found_files:
        print("Archivos encontrados:")
        for file in found_files:
            print(file)
    else:
        print("No se encontraron archivos.")
```
