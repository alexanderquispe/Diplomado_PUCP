```python
# We import the necessary libraries
import pandas as pd
import os
import re
```

```python
#We define the name of the function and attributes
def import_data_sbs_banks(start_month, start_year, end_month=None, end_year=None):
    path = r"../../_data/sbs/B_RawData/bancos"  # Assign the folder path
    
    if end_month is None:
        end_month = start_month  # Use start_month if end_month isn't provided
    if end_year is None:
        end_year = start_year  # Use start_year if end_year isn't provided

    data_month_year = {}

    # Assign the standard name of the Excel files obtained from the SBS data scraping
    patron = re.compile(fr"table_clean_(\d+)_(\d+)_{start_year}")
    
    #We create a loop to iterate through the files in the folder and select those Excel files
    for file in os.listdir(path):
        if file.endswith('.xlsx') and patron.match(file):
            partes_nombre = file.split('_')

            # Extracting the day, month and year from the file name
            match = patron.match(file)
            day = int(match.group(1))
            month = int(match.group(2))
            year = start_year

            # Checking if the file complies with the date data
            if start_year <= year <= end_year and start_month <= month <= end_month:
                file_path = os.path.join(path, file)
                df = pd.read_excel(file_path)

                # Storing data in the dictionary
                if year not in data_month_year:
                    data_month_year[year] = {}

                if month not in data_month_year[year]:
                    data_month_year[year][month] = df
                else:
                    # If there is already data for that month, concatenate the data for that month
                    data_month_year[year][month] = pd.concat(
                        [data_month_year[year][month], df], ignore_index=True)
    # We create an additional conditional to indicate when the dictionary has been created successfully
    if data_month_year:
        print("Dictionary successfully created for the given month and years")
    else:
        print("No data matching the selected month and year were found")
    
    return data_month_year
```

```python
# You can replace the dates in the following arguments to import the data, remember that if you don't have the dates of the end of the month and year, you can put None where it corresponds.
start_month = 2
start_year = 2009
end_month = None
end_year = None

import_data_sbs_banks(start_month, start_year, end_month, end_year)
```
