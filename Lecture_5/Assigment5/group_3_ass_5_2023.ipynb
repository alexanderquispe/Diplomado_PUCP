1. Generate a function called **import_data_sbs_banks** that take as arguments `start_month`, `start_year`, `end_month` and `end_year`. This function should be able to import the files in **..\_data\sbs\B_RawData\bancos** taking into consideration these arguments. Additionally, it should work if it is only given the first two arguments `(start_mont, start_year)`. The function should `returns` a nested dictionary. The `main key` must be year and `second key` must be `month`. The values should be the dataframes.

```python
import os
import pandas as pd

def importar_datos_sbs_bancos(inicio_mes, inicio_anio, fin_mes=None, fin_anio=None):
    
    # Definir la ruta de los datos
    ruta_datos = r"../../_data/sbs/B_RawData/bancos"
    
    # Listar todos los archivos en la ruta de datos
    archivos = os.listdir(ruta_datos)
    
    # Diccionario donde se almacenarán los datos
    diccionario_resultado = {}
    
    for archivo in archivos:
        # Verificar si el archivo comienza con "table_clean_" y tiene extensión .xlsx
        if archivo.startswith("table_clean_") and archivo.endswith(".xlsx"):
            # Si es así, extraer día, mes y año del nombre del archivo
            _, _, dia, mes, anio = archivo[:-5].split('_')
            # Convertir a enteros lo necesario (mes y año)
            mes = int(mes)
            anio = int(anio)
            
            # Verificar si los valores en 'inicio_anio', 'inicio_mes', 'fin_anio' y 'fin_mes' del archivo
            # se encuentran dentro del rango de tiempo especificado (rango definido al usar la función)
            if (inicio_anio < anio or (inicio_anio == anio and inicio_mes <= mes)) and \
               (fin_anio is None or (fin_anio > anio or (fin_anio == anio and fin_mes >= mes))):
                
                try:
                    # Si los valores están en el rango, construir la ruta completa del archivo
                    ruta_archivo = os.path.join(ruta_datos, archivo)
                    # Leer el archivo Excel en un DataFrame
                    df = pd.read_excel(ruta_archivo)
                    
                    # Verificar si la clave de año existe en el diccionario_resultado, si no, crearla
                    if anio not in diccionario_resultado:
                        diccionario_resultado[anio] = {}
                    
                    # Almacenar el DataFrame en diccionario_resultado bajo el año y mes correspondientes
                    diccionario_resultado[anio][mes] = df
                
                # Manejar el caso en el que el archivo no se encuentra
                except FileNotFoundError:
                    print(f"El archivo {archivo} no se encuentra.")
                   
                
    # Ordenar diccionario_resultado por año y mes
    diccionario_ordenado = dict(sorted(diccionario_resultado.items()))

    for anio, meses in diccionario_ordenado.items():
        diccionario_ordenado[anio] = dict(sorted(meses.items()))
    
    # Devolver el diccionario resultante
    return diccionario_ordenado
```

Example 1: To import all the files in data path for a range

```python
# Primero, definimos el rango: por ejemplo, la segunda mitad del año 2008, momentos críticos de la crisis financiera de 2008
inicio_mes = 7
inicio_anio = 2008
fin_mes = 12
fin_anio = 2008

# Segundo, utilizamos la función (corregir el nombre de la función aquí)
datos_dict = importar_datos_sbs_bancos(inicio_mes, inicio_anio, fin_mes, fin_anio)

# Finalmente, imprimimos los resultados
print(datos_dict)
```
