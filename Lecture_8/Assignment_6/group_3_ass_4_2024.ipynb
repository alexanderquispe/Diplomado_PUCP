```python
def get_coordinates(address):
    #api_key = "AIzaSyCbLkEjRx999HGozo7ond7jScp9cQ4vUJo" #API Used: Geocoding. This is Daniela's key. The next person should use their own key
    api_key = "AIzaSyAdFmDmnVPmqMZy6NPHWVOQ7b_VqIHUPU4" #API Used: Geocoding. This is Guadalupe's key. The next person should use their own key
#   api_key = "AIzaSyCt3RSeThYlA7H_PQUO-caYKvz4N52Wc_M" #API Used: Geocoding. This is Luis's key.
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    # Replace spaces with '+' and encode the address
    formatted_address = "+".join(address.split())
    
    # Construct the URL for the API request
    url = f"{base_url}?address={formatted_address}&key={api_key}&language=es&region=pe"

    # Make the API request
    response = requests.get(url)

    # Parse the JSON response
    data_json = response.json()

    # Extract latitude and longitude if the request was successful
    if data_json['status'] == 'OK':
        location = data_json['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        return None, None

# Iterate through rows in the DataFrame and update 'Lat' and 'Lng' columns
df['Lat'] = None
df['Lng'] = None

for index, row in df.iterrows():
    direccion = row['Direccion']
    lat, lng = get_coordinates(direccion)

    # Update the DataFrame with the obtained coordinates
    df.at[index, 'Lat'] = lat
    df.at[index, 'Lng'] = lng

# Display the updated DataFrame
print("Updated DataFrame:")
print(df)
```

```python
# Function to obtain driving time and distance from BBVA establishments to our addresses
def get_driving_info(origin_lat, origin_lng, dest_lat, dest_lng, api_key):
    base_url = "https://maps.googleapis.com/maps/api/directions/json"

    # Construct the URL for the API request
    url = f"{base_url}?origin={origin_lat},{origin_lng}&destination={dest_lat},{dest_lng}&key={api_key}&region=es"

    # Make the API request
    response = requests.get(url)

    # Parse the JSON response
    data_json = response.json()

    # Extract driving time and distance if the request was successful
    if data_json['status'] == 'OK':
        route = data_json['routes'][0]['legs'][0]
        driving_time = route['duration']['text']
        driving_distance = route['distance']['text']
        return driving_time, driving_distance
    else:
        return None, None

# Create a list of tuples with our names and coordinates
participants = [("Daniela", LatDaniela, LngDaniela), ("Diego", LatDiego, LngDiego), ("Guadalupe", LatGuadalupe, LngGuadalupe), ("Luis", LatLuis, LngLuis), ("Alvaro", LatAlvaro, LngAlvaro)]

# A loop is created to iterate in the dataframe and the columns are updated
for participant, origin_lat, origin_lng in participants:
    df[f'DrivingTime{participant}'] = None
    df[f'DrivingDistance{participant}'] = None

    for index, row in df.iterrows():
        lat = row['Lat']
        lng = row['Lng']

        # Get time and distance information
        driving_time, driving_distance = get_driving_info(origin_lat, origin_lng, lat, lng, "AIzaSyAdFmDmnVPmqMZy6NPHWVOQ7b_VqIHUPU4") #This is Guadalupe's key. The next person should use their own key

        # Update the DataFrame with the obtained driving time and distance.
        df.at[index, f'DrivingTime{participant}'] = driving_time
        df.at[index, f'DrivingDistance{participant}'] = driving_distance

df
```
