```python
# Function to get driving time and distance from San Diego to a location
def get_driving_info(lat, lng):
    api_key = "AIzaSyCbLkEjRx999HGozo7ond7jScp9cQ4vUJo" #API Used: Routes API. This is Daniela's key. The next person should use their own key
    base_url = "https://maps.googleapis.com/maps/api/directions/json"

    # Construct the URL for the API request
    url = f"{base_url}?origin={LatDaniela},{LngDaniela}&destination={lat},{lng}&key={api_key}&region=es"

    # Make the API request
    response = requests.get(url)

    # Parse the JSON response
    data_json = response.json()

    # Extract driving time and distance if the request was successful
    if data_json['status'] == 'OK':
        route = data_json['routes'][0]['legs'][0]
        driving_time = route['duration']['text']
        driving_distance = route['distance']['text']
        return driving_time, driving_distance
    else:
        return None, None

# Iterate through rows in the DataFrame and update 'DrivingTime' and 'DrivingDistance' columns
df['DrivingTimeDaniela'] = None
df['DrivingDistanceDaniela'] = None

for index, row in df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    
    driving_time, driving_distance = get_driving_info(lat, lng)

    # Update the DataFrame with the obtained driving time and distance
    df.at[index, 'DrivingTimeDaniela'] = driving_time
    df.at[index, 'DrivingDistanceDaniela'] = driving_distance

    
df    
```
