### 1. Take the information from national and private universities.

```python
driver = webdriver.Chrome( ChromeDriverManager().install() )
driver.maximize_window()
```

```python
url = 'https://www.sunedu.gob.pe/lista-universidades/'
driver.get( url )
```

#### Public

```python
public_type = driver.find_element_by_xpath( '/html/body/div[1]/div[3]/div/div/div[1]/div/div/div/div[1]/ul/li[1]/a' )
public_type.click()
```

```python
table_public = driver.find_element_by_id( "tablepress-22_wrapper" )
```

```python
table_public_html = table_public.get_attribute ('innerHTML')
pd_table_public = pd.read_html (table_public_html)
```

```python
format_table_public = pd_table_public [0]
format_table_public
```

#### Private

```python
private_type = driver.find_element_by_xpath( '/html/body/div[1]/div[3]/div/div/div[1]/div/div/div/div[1]/ul/li[2]/a' )
private_type.click()
```

```python
table_private = driver.find_element_by_id( "tablepress-23_wrapper" )
```

```python
table_private_html = table_private.get_attribute ( 'innerHTML' )
pd_table_private = pd.read_html ( table_private_html )
```

```python
format_table_private = pd_table_private [0]
format_table_private
```

```python
univ_append = pd.concat([format_table_public, format_table_private]).reset_index(drop = True)
univ_append.rename(columns = {'DEPARTAMENTO': 'Department_univ', 'PROVINCIA': 'Province_univ', 'UNIVERSIDAD': 'Name_univ'}, inplace = True)
```

```python
data_univ = univ_append[['Department_univ', 'Province_univ', 'Name_univ']].assign(Adress_univ = univ_append.Name_univ + ', ' + univ_append.Province_univ + ', ' + univ_append.Department_univ)
data_univ.head()
```
